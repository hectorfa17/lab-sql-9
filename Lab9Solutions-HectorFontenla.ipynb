**Actual Lab**

```python
#How many distinct (different) actors' last names are there?

query = '''SELECT count(distinct(last_name)) distict_last_name_count FROM actor;'''

dist_actors = pd.read_sql_query(query, engine)
dist_actors.head()
```

```python
#Get all films with ARMAGEDDON in the title.

query = '''SELECT title
FROM film
WHERE title LIKE '%%ARMAGEDDON%%';'''

arma = pd.read_sql_query(query, engine)
arma.head(10)
```

```python
#How many films include Behind the Scenes content?

query = '''SELECT count(special_features)
FROM film
WHERE special_features LIKE '%%behind the scenes%%';'''

spefea = pd.read_sql_query(query, engine)
spefea.head(10)
```

```python
#Which kind of movies (rating) have a mean duration of more than two hours?

query ='''SELECT rating, AVG(ROUND(length, 2)) as avg_time FROM film
GROUP BY rating
HAVING avg_time > 120;'''

avglen = pd.read_sql_query(query, engine)
avglen.head(10)
```

```python
#Rank films by length (filter out the rows that have nulls or 0s in length column). 
#In your output, only select the columns title, length, and the rank.

query = '''SELECT title, length 
FROM film 
WHERE length IS NOT NULL OR length <> 0 
GROUP BY film_id ORDER BY length DESC;'''

films = pd.read_sql_query(query, engine)
films.head(10)
```
